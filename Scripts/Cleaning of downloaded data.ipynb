```python
import pandas as pd

# Read the data from the sheet into a DataFrame.
df = pd.read_excel("GDP.xls", sheet_name='Data')
df
```

```python
columns_to_keep = ["Country Name", "2021", "2022"]

# Select only the desired columns using .loc
df = df.loc[:, columns_to_keep]
df
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
df = df.rename(columns={'Country Name': 'Country'})
df = df.rename(columns={'2021': 'GDP 2021'})
df = df.rename(columns={'2022': 'GDP 2022'})
df
```

```python
csv_file = "R_GDP.csv"
df.to_csv(csv_file, index=False, encoding='utf-8')
print(f"DataFrame saved to {csv_file}")
```

```python
# Read the data from the sheet into a DataFrame.
df = pd.read_excel("population.xls", sheet_name='Data')
df = df.rename(columns={'Country Name': 'Country'})
df
```
