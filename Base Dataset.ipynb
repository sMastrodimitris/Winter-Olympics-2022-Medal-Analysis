```python
import pandas as pd
import glob

# Step 1: Load the CSV files into separate DataFrames
csv_files = glob.glob('Cleaned Data/*.csv')  
dataframes = [pd.read_csv(file) for file in csv_files]

# Step 2: Merge DataFrames based on the "Country" column
merged_dataframe = dataframes[0]
for df in dataframes[1:]:
    # Perform an outer merge on the "Country" column to retain all data
    merged_dataframe = pd.merge(merged_dataframe, df, on="Country", how="outer")

# Display the resulting merged DataFrame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
merged_dataframe.head(200)
```

```python
# Assuming 'merged_dataframe' is your DataFrame
threshold = len(merged_dataframe.columns) - 5
filtered_dataframe = merged_dataframe.dropna(thresh=threshold)
filtered_dataframe = filtered_dataframe.reset_index(drop=True)
filtered_dataframe = filtered_dataframe.drop('GDP 2021', axis=1)

filtered_dataframe['GDP 2022'] = filtered_dataframe['GDP 2022'].apply(lambda x: '{:.2f}'.format(x))
filtered_dataframe['population 2022'] = filtered_dataframe['population 2022'].apply(lambda x: '{:.2f}'.format(x))


filtered_dataframe['population 2022'] = filtered_dataframe['population 2022'].astype(float)
filtered_dataframe['population 2022'] = filtered_dataframe['population 2022'].apply(lambda x: int(x) if x.is_integer() else x)


# Display the filtered DataFrame
filtered_dataframe.head(30)
```

```python
filtered_dataframe.loc[20, 'Population Density'] = 133
filtered_dataframe.loc[20, 'Urbanization (%)'] = 74.1
filtered_dataframe.head(30)
```

```python
filtered_dataframe['Order'] = filtered_dataframe['Order'].astype(int)
filtered_dataframe['Gold'] = filtered_dataframe['Gold'].astype(int)
filtered_dataframe['Silver'] = filtered_dataframe['Silver'].astype(int)
filtered_dataframe['Bronze'] = filtered_dataframe['Bronze'].astype(int)
filtered_dataframe['Total'] = filtered_dataframe['Total'].astype(int)
filtered_dataframe['Order by Total'] = filtered_dataframe['Order by Total'].astype(int)
filtered_dataframe['Elevation'] = filtered_dataframe['Elevation'].astype(int)
filtered_dataframe['Population Density'] = filtered_dataframe['Population Density'].astype(int)

filtered_dataframe.head(30)
```

```python
import math
filtered_dataframe['Average Yearly Temperature'] = filtered_dataframe['Average Yearly Temperature'].apply(lambda x: math.ceil(float(x.replace('−', '-')))).astype(int)
filtered_dataframe.head(30)
```

```python
filtered_dataframe['Average height men'] = filtered_dataframe['Average height men'].apply(lambda x: float(x.replace(' m', '')))
filtered_dataframe['Average height women'] = filtered_dataframe['Average height women'].apply(lambda x: float(x.replace(' m', '')))
filtered_dataframe['Weight men'] = filtered_dataframe['Weight men'].apply(lambda x: float(x.replace(' kg', '')))
filtered_dataframe['Weight women'] = filtered_dataframe['Weight women'].apply(lambda x: float(x.replace(' kg', '')))
filtered_dataframe.head(30)
```

```python
filtered_dataframe['GDP 2022'] = filtered_dataframe['GDP 2022'].apply(lambda x: int(x[:-3]))
filtered_dataframe['gdp_per_capita'] = filtered_dataframe['GDP 2022'] / filtered_dataframe['population 2022']
filtered_dataframe['gdp_per_capita'] = filtered_dataframe['gdp_per_capita'].astype(int)
filtered_dataframe['Urbanization (%)'] = filtered_dataframe['Urbanization (%)'].astype(float) / 100
filtered_dataframe.head(30)                                                                      
```

```python
filtered_dataframe = filtered_dataframe.reindex(columns=['Order','Order by Total', 'Country Code', 'Country','Gold', 'Silver', 'Bronze','Total', 'Elevation', 'Average Yearly Temperature','GDP 2022','gdp_per_capita','population 2022', 'Average height men','Weight men', 'BMI men', 'Average height women','Weight women', 'BMI women', 'Population Density', 'Urbanization (%)'])
filtered_dataframe.head(30)

```

```python
# Assuming 'filtered_dataframe' is your DataFrame
output_file_name = "filtered_data.xlsx"

# Save the DataFrame to a temporary Excel file
temp_file_name = "temp_filtered_data.xlsx"
filtered_dataframe.to_excel(temp_file_name, index=False)

# Rename the temporary file to the original file name
import os
#os.rename(temp_file_name, output_file_name)

print(f"DataFrame saved to {output_file_name}")
```
