# Statistical Analysis

#### Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

#### Loading and exploring the data

```python
df = pd.read_excel("temp_filtered_data.xlsx")
df
```

```python
df.info()
```

#### Descriptive Statistics

```python
start_col = 4
end_col = 21

cols_to_process = df.iloc[:, start_col:end_col+1]
result = cols_to_process.agg(['mean', 'median', 'std', 'max', 'min'])
range_result = cols_to_process.max() - cols_to_process.min()
result = result.append(range_result.rename('range'))

print(result)
```

#### Top performers for Gold

```python
dfplot = df.sort_values(by='Gold', ascending=False).head(5)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Gold', data=dfplot)
plt.xlabel('Country')
plt.ylabel('Gold')
plt.title('Top 5 Countries by Gold')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

#### Top performers for Silver

```python
dfplot = df.sort_values(by='Silver', ascending=False).head(5)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Silver', data=dfplot)
plt.xlabel('Country')
plt.ylabel('Silver')
plt.title('Top 5 Countries by Silver')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

#### Exploring Correlations for Team Size, GDP and GDP per Capita

```python
plt.figure(figsize=(10, 6))
scatterplot = sns.regplot(x='Total', y='Team Size', data=df, scatter_kws={'s': 100})


for index, row in df.iterrows():
    scatterplot.text(row['Total'], row['Team Size'], row['Country'], ha='center', va='bottom')

plt.xlabel('Total')
plt.ylabel('Team Size')
plt.title('Total number of Medals and Team Size Scatter Plot')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatterplot = sns.regplot(x='Total', y='GDP 2022', data=df, scatter_kws={'s': 100}) 


for index, row in df.iterrows():
    scatterplot.text(row['Total'], row['GDP 2022'], row['Country'], ha='center', va='bottom')

plt.xlabel('Total')
plt.ylabel('GDP 2022')
plt.title('Total number of Medals and GDP Scatter Plot')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatterplot = sns.regplot(x='Total', y='gdp_per_capita', data=df, scatter_kws={'s': 100}) 


for index, row in df.iterrows():
    scatterplot.text(row['Total'], row['gdp_per_capita'], row['Country'], ha='center', va='bottom')

plt.xlabel('Total')
plt.ylabel('GDP per Capita')
plt.title('Total number of Medals and GDP per Capita Scatter Plot')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

#### Exploring Correlation for the economic and climate factors

```python
selected_columns = ['Gold', 'Total','Average Yearly Temperature', 'Elevation',
                    'GDP 2022', 'gdp_per_capita', 'population 2022','Team Size']

sns.set(style="ticks")
sns.pairplot(df[selected_columns], kind="reg")
plt.suptitle("Pairplot with Correlation Lines", y=1.02) 

plt.show()
```

```python
# Select specific columns for the pairplot
selected_columns = ['Gold', 'Silver', 'Bronze', 'Total','Average Yearly Temperature', 'Elevation',
                    'GDP 2022', 'gdp_per_capita', 'population 2022','Team Size']

# Calculate correlation matrix
correlation_matrix = df[selected_columns].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

plt.title("Correlation Heatmap")
plt.show()
```

#### Applying appropriate metrics for temperature

```python
# Select specific columns for the correlation analysis
selected_columns = ['Gold', 'Silver', 'Bronze', 'Total', 'Average Yearly Temperature', 'Country']

# Calculate Pearson and Spearman correlation coefficients
pearson_corr = df[selected_columns[:-1]].corr(method='pearson')
spearman_corr = df[selected_columns[:-1]].corr(method='spearman')


print("Pearson Correlation:\n", pearson_corr)
print("\nSpearman Correlation:\n", spearman_corr)
```

#### Plotting for Gold, Silver and Bronze

```python
# Set style parameters
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

scatterplot = sns.scatterplot(x='Average Yearly Temperature', y='Gold', data=df, linewidth=5, edgecolor='k')
sns.kdeplot(data=df, x='Average Yearly Temperature', y='Gold', ax=scatterplot, cmap='Blues', fill=True, levels=11, alpha=0.8)
plt.xlabel('Average Yearly Temperature')
plt.ylabel('Gold')
plt.title('Scatter Plot of Average Yearly Temperature for Gold Medals')


for index, row in df.iterrows():
    scatterplot.text(row['Average Yearly Temperature'], row['Gold'], row['Country'], ha='center', va='bottom', fontsize=16)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
scatterplot = sns.scatterplot(x='Average Yearly Temperature', y='Silver', data=df, linewidth=5, edgecolor='k')
sns.kdeplot(data=df, x='Average Yearly Temperature', y='Silver', ax=scatterplot, cmap='Blues', fill=True, levels=11, alpha=0.8)
plt.xlabel('Average Yearly Temperature')
plt.ylabel('Silver')
plt.title('Scatter Plot of Average Yearly Temperature for Silver Medals')


for index, row in df.iterrows():
    scatterplot.text(row['Average Yearly Temperature'], row['Silver'], row['Country'], ha='center', va='bottom', fontsize=16)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
scatterplot = sns.scatterplot(x='Average Yearly Temperature', y='Bronze', data=df, linewidth=2, edgecolor='k')
sns.kdeplot(data=df, x='Average Yearly Temperature', y='Bronze', ax=scatterplot, cmap='Blues', fill=True, levels=11, alpha=0.8)
plt.xlabel('Average Yearly Temperature')
plt.ylabel('Bronze')
plt.title('Scatter Plot of Average Yearly Temperature for Bronze Medals')

for index, row in df.iterrows():
    scatterplot.text(row['Average Yearly Temperature'], row['Bronze'], row['Country'], ha='center', va='bottom', fontsize=16)

plt.show()
```
